```
import wave
import numpy as np
import matplotlib.pyplot as plt

wf = wave.open("aiueo_wavesurfer.wav" , "r" )
buf = wf.readframes(wf.getnframes())
# バイナリデータを16bit整数に変換
data = np.frombuffer(buf, dtype="int16")
plt.plot(data[500:10000])
plt.show()          # グラフ表示
```
