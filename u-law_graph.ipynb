### u-law

```python
%matplotlib inline
 
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-1, 1, 0.01)
y=x*np.log2(1+225*np.abs(x))/np.log2(1+225.0)

plt.title("u-law")
plt.xlabel("input")
plt.ylabel("output")
plt.xlim([-1.0,1.0])
plt.ylim([-1.0,1.0])
plt.scatter(x,y)
plt.show()
#plt.savefig("img.png")
```

### A-law

```python
%matplotlib inline
 
import numpy as np
import matplotlib.pyplot as plt
A=87.7
x=np.arange(-1, 1, 0.01)
y=np.where(np.abs(x)<1/A, x*A*np.abs(x)/(1+np.log2(A)), x*(1+np.log2(A*np.abs(x)))/(1+np.log2(A)))

plt.title("A-law")
plt.xlabel("input")
plt.ylabel("output")
plt.xlim([-1.0,1.0])
plt.ylim([-1.0,1.0])
plt.scatter(x,y)
plt.show()
#plt.savefig("img.png")
```
