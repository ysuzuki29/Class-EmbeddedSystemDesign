### ocv_image_convert

```python
# -*- coding: utf-8 -*-
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = cv2.imread("picture_from_cam.jpg")

#print(img)
print(img[200,200])
print(img[320,200])

cv2.imwrite("picture_from_cam.png", img)
cv2.imwrite("picture_from_cam.bmp", img)

img_list=np.asarray(img)
plt.imshow(img_list)
plt.show()
```

```python
# -*- coding: utf-8 -*-
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = cv2.imread("picture_from_cam.bmp")
print("img.shape:  ", img.shape)
gray = cv2.imread("picture_from_cam.bmp", 0)
print("gray.shape: ", gray.shape )

img_list=np.asarray(img)
plt.axhline(200)
plt.axvline(280)

plt.imshow(img_list)
plt.show()
```
