Digistsの表示

```
# -*- coding: utf-8 -*-
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

#データの読み込み
#引数　2〜10 引数無し→10
digits=load_digits()

#データの構造
#画像数:1797, 各画像のpixel数:8x8=64
print(digits.data.shape)

#データを画像で確認
for j in range(0, 20):
    print("Label:", digits.target[j])
    print(digits.images[j])
    plt.imshow(digits.images[j], cmap="gray", interpolation='NONE')
    plt.show()
```
