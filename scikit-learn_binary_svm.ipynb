```
# -*- coding: utf-8 -*-
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC, SVC
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score
import numpy as np
import matplotlib.pyplot as plt

#データの読み込み
#引数2-10 引数無し→10
digits = load_digits()

#データを図で確認
for j in range(0, 10):
    print(digits.target[j])
    print(digits.images[j])
    plt.imshow(digits.images[j], cmap="gray", interpolation='NONE')
    plt.show()    

# label:0-9 -> 5 or 0    
digits.target=[1 if digits.target[t] == 5 else 0 for t in range(len(digits.target))]
digits.target=np.array(digits.target)
    
#トレーニングデータとテストデータに分割
data_train, data_test, label_train, label_test = train_test_split(digits.data, digits.target, test_size=0.25)

#データを確認
print("training")
print(label_train.shape)
for i in range(0, 10):
    #ラベル
    print(label_train[i],)
    #データ
    print(data_train[i])

print("test")
print(label_test.shape)
for i in range(0, 10):
    #ラベル
    print(label_test[i],)
    #データ
    print(data_test[i])

#分類器にパラメータを与える
estimator = LinearSVC(C=1.0)
#estimator = SVC(C=1.0, kernel='rbf', gamma='auto')

#トレーニングデータで学習
estimator.fit(data_train, label_train)

#テストデータの予測
label_predict = estimator.predict(data_test)

#結果と比較
print("\nactual class")
print(label_test)
print("\npredicted class")
print(label_predict)

#confusion matrix
print("\nconfusion matrix")
print("predicted class")
print(confusion_matrix(label_test, label_predict))

conf_mx=confusion_matrix(label_test, label_predict)
fn=conf_mx[0,0]
fp=conf_mx[0,1]
tn=conf_mx[1,0]
tp=conf_mx[1,1]

print("fn=%d fp=%d tn=%d tp=%d" % (fn, fp, tn, tp))

# confusion matrix 
print("confusion matrix: ")
plt.matshow(conf_mx, cmap=plt.cm.gray)
plt.show()

# confusion matrix (filling 0 in diagonal and normarization)
row_sums = conf_mx.sum(axis=1, keepdims=True)
norm_conf_mx = conf_mx / row_sums
np.fill_diagonal(norm_conf_mx, 0)
plt.matshow(norm_conf_mx, cmap=plt.cm.gray)
plt.show()

# accuracy
print("accuracy")
print("accuracy=%f" % accuracy_score(label_test, label_predict))
print("accuracy=%f")

# recall
print("recall")
print("recall=%f" % recall_score(label_test, label_predict))
print("recall=%f")
#rec=

# precision
print("precision")
print("precision=%f" % precision_score(label_test, label_predict))
print("precision=%f")
#prec=

# f-measure
print("f-measure")
print("f1_score=%f" % f1_score(label_test, label_predict))
print("f1_score=%f")
print("f2_score=%f")
print("f0.5_score=%f")
```
