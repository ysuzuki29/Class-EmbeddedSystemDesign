### macro/micro average precision/recall/F-score

```python
# -*- coding: utf-8 -*-
from sklearn.datasets import load_digits
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC, SVC
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

#
#2-10 10
digits = load_digits()

#
print("Digit data (first 10 data)")
for j in range(0, 10):
    print(digits.target[j])
    print(digits.images[j])
    plt.imshow(digits.images[j], cmap="gray", interpolation='NONE')
    plt.show()

# test_size=0.25 -> training data:3/4, test data:1/4
data_train, data_test, label_train, label_test = train_test_split(digits.data, digits.target, test_size=0.25)

#
print("\ntraining data (first 10 data)")
for i in range(0, 10):
    #
    print(label_train[i],)
    #
    print(data_train[i])

print("\ntest data (first 10 data)")
for i in range(0, 10):
    #
    print(label_test[i])
    #
    print(data_test[i])

#
estimator = LinearSVC(C=1.0)
#estimator = SVC(C=1.0, kernel='rbf', gamma='auto')

#
estimator.fit(data_train, label_train)

#
label_predict = estimator.predict(data_test)

#comparison between actual class and predicted class 
print("\nanswer")
print(label_test)
print("\npredict")
print(label_predict)

#confusion matrix
print("\nconfusion matrix")
from sklearn.metrics import confusion_matrix
print(confusion_matrix(label_test, label_predict))

conf_mx=confusion_matrix(label_test, label_predict)

# confusion matrix 
print("confusion matrix: ")
plt.matshow(conf_mx, cmap=plt.cm.gray)
plt.show()

# confusion matrix (filling 0 in diagonal and normarization)
# 
#
row_sums = conf_mx.sum(axis=1, keepdims=True)
norm_conf_mx = conf_mx / row_sums
np.fill_diagonal(norm_conf_mx, 0)
plt.matshow(norm_conf_mx, cmap=plt.cm.gray)
plt.show()

#macro average precision 
print("\nprecision macro")
print(precision_score(label_test, label_predict, average='macro'))

#micro average precision 
print("\nprecision micro")
print(precision_score(label_test, label_predict, average='micro'))

#macro average recall 
print("\nrecall macro")
print(recall_score(label_test, label_predict, average='macro'))

#micro average recall 
print("\nrecall micro")
print(recall_score(label_test, label_predict, average='micro'))

#macro average F-score F
print("\nf1 macro")
print(f1_score(label_test, label_predict, average='macro'))

#micro average F-score F
print("\nf1 micro")
print(f1_score(label_test, label_predict, average='micro'))
```
