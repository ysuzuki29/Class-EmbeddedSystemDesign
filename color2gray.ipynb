```
# -*- coding: utf-8 -*-
# usage: color2gray.py colorpicture(without .jpg) bin_thresh
import cv2
import sys

fname="picture_from_cam"

colorfilename=fname+".jpg"
grayfilename=fname+"G.jpg"
binfilename=fname+"B"+sys.argv[2]+".jpg"
thresh=64

# 元の画像
img_src = cv2.imread(colorfilename, 1)

# グレースケール
img_dst = cv2.cvtColor(img_src, cv2.COLOR_BGR2GRAY)
cv2.imwrite(grayfilename, img_dst)

# 二値画像
ret, img_bin = cv2.threshold(img_dst, thresh, 255, cv2.THRESH_BINARY)
cv2.imwrite(binfilename, img_bin)
# 
print("元の画像")
print(img_src)
print("次元ごとの配列の大きさ")
print(img_src.shape)
print()

# 
print("grayscale変換後の画像")
print(img_dst)
print("次元ごとの配列の大きさ")
print(img_dst.shape)
print()

# 
print("二値画像変換後の画像")
print(img_bin)
print("次元ごとの配列の大きさ")
print(img_bin.shape)
```
