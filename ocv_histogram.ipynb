```
# -*- coding: utf-8 -*-
import cv2
import math
import numpy as np
from google.colab.patches import cv2_imshow

#img_src= cv2.imread("picture_from_cam.jpg", 1)
#img_src= cv2.imread("picture_from_cam2.jpg", 1)
#img_src= cv2.imread("kogyokai.jpg", 1)
img_src= cv2.imread("kogyokaiG.jpg", 0)
#img_src= cv2.imread("kogyokaiB100.jpg", 0)

#cv2_imshow(img_src)

img_dst=np.zeros([100, 256]).astype('uint8')
rows, cols = img_dst.shape

size=(256, 100)
img_small=cv2.resize(img_src, size)
cv2_imshow(img_small)
transpose_img=img_small.transpose(1,0)
clockwise_img=transpose_img[:,::-1]
counter_clockwise_img=transpose_img[::-1]
cv2_imshow(clockwise_img)
cv2_imshow(counter_clockwise_img)

#
hdims=[256]
hranges=[0, 256]
hist=cv2.calcHist([img_src], [0], None, hdims, hranges)

min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(hist)

for i in range(0, 255):
    v=hist[i]
    cv2.line(img_dst, (i, rows), (i, rows-rows*(v/max_val)), (255, 255, 255))
cv2_imshow(img_dst)


#while True:
#    k=cv2.waitKey(1)
#    if k==27: # ESC key
#        break
#
#cv2.destroyAllWindows()
```
